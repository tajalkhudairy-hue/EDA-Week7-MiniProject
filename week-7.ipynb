```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
path="/kaggle/input/datasets/tajmomaaaa/day-35/day35_project.csv"
df = pd.read_csv(path)
df.head()
plt.figure()
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure()
sns.boxplot(x='segment', y='income', data=df)
plt.title("Income by Segment")
plt.show()

plt.figure()
sns.scatterplot(x='income', y='spend', hue='segment', data=df)
plt.title("Income vs Spend")
plt.show()

plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

print("""
Insights:
1. Higher income is associated with higher spending.
2. Segment C shows higher income levels.
3. Age has a moderate influence on spending.
""")

print("""
Recommendations:
1. Use income as a key feature.
2. Use segment for customer behavior analysis.
3. Age can be used as a supporting feature.
""")
```
